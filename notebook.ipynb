<h3>Data cleaning</h3>

```python
# here we will convert the numeric columns in right Datatypes
df['discounted_price']  = df['discounted_price'].str.replace('₹' ,'')
df['discounted_price'] = df['discounted_price'].str.replace(',' , '').astype('float')

df['actual_price'] =df['actual_price'].str.replace('₹','')
df['actual_price'] = df['actual_price'].str.replace(',' , '').astype('float') 

df['rating'] = round(df['rating'].str.replace('|' , '0').astype('float') ,1)  
df['discount_percentage'] = pd.to_numeric(df['discount_percentage'].str.replace('%' ,''))
df['rating_count'] = pd.to_numeric(df['rating_count'].str.replace(',' ,''))
```

```python
def extract_first_category(row):
        categories = row.split('|')
        first_category = categories[0].split('&')[0]  # Extracting the first category before '&'
        return first_category
```

```python
df['main_category'] = df['category'].apply(extract_first_category)
df['main_category'].value_counts()
```

```python
import pandas as pd
def extract_second_category(row):
    if '&' in row:  # If '&' exists, extract the segment between '&' and '|'
        second_category_segment = row.split('&')[1].split('|')[0]
    else:  # If '&' doesn't exist, there is no second category segment
        second_category_segment = None
    return second_category_segment

df['second_category'] = df['category'].apply(extract_second_category)
df['second_category'].value_counts()
```

```python
def extract_last_category(row):
    categories = row.split('|')
    last_category = categories[-1] if categories[-1] else None
    return last_category

# Apply the function to each row of the 'category' column
df['type_of_product'] = df['category'].apply(extract_last_category)
df['type_of_product'].value_counts()
```

```python
df = df.drop(['product_link' , 'img_link'] , axis = 1)
df.columns
```

<h4>the Top 20 most common word in Product description ? </h4>

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import re
from nltk.corpus import stopwords
import matplotlib.pyplot as plt

# Assuming 'product_descriptions' is the column containing the text data
product_descriptions = df_new['about_product']

# Define a function to preprocess the text data
def preprocess_text(text):
    # Remove punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Convert to lowercase
    text = text.lower()
    # Remove stop words
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    # Join the words back into a single string
    text = ' '.join(filtered_words)
    return text

# Preprocess the text data
product_descriptions = product_descriptions.apply(preprocess_text)

# Tokenization and word frequency calculation
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(product_descriptions)
word_count_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Calculate word frequencies
word_frequencies = word_count_df.sum().sort_values(ascending=False)

# Visualize the most frequent words
top_words = word_frequencies.head(20)
top_words.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Top 20 Most Frequent Words in Product Descriptions')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()

# the words that attract customers are these :
# 1- usb customers focus on products that have usb like : usb cable , tv have usb , chraging have usb
# 2- customers attract products that have word cable maybe : cable usb , powerbank with cable 
# 3- charging : charging with usb , labtop with high power of charging 
# 4- power : like power supply , power bank , labtop with high power charing , 
# and soon on
```

<h4> what aspects of the products customers are most interested in or concerned about ?</h4>

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation


# Assuming 'review_title' is the column containing the review titles
review_titles = df_new['review_title']

# Vectorize the text data
vectorizer = CountVectorizer(stop_words='english')
X = vectorizer.fit_transform(review_titles)

# Define the number of topics
num_topics = 5  # Adjust the number of topics as needed

# Apply LDA
lda = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda.fit(X)

# Display the topics and the top words in each topic
feature_names = vectorizer.get_feature_names_out()
topic_words = {}
for topic_idx, topic in enumerate(lda.components_):
    top_words_idx = topic.argsort()[:-11:-1]  # Get indices of top 10 words
    top_words = [feature_names[i] for i in top_words_idx]
    topic_words[f"Topic {topic_idx + 1}"] = top_words

# Display the topics and top words
for topic, words in topic_words.items():
    print(f"{topic}: {', '.join(words)}")
```

```python
def categorize_nps(rating):
    if rating >= 4.0:
        return 'Promoter'
    elif rating >= 3.0:
        return 'Passive'
    else:
        return 'Detractor'

# Apply NPS categorization to ratings
df_new['nps_category'] = df_new['rating'].apply(categorize_nps)

# Calculate NPS
nps_counts = df_new['nps_category'].value_counts(normalize=True)
nps_score = (nps_counts['Promoter'] - nps_counts['Detractor']) * 100

print("NPS Score:", nps_score)
print("Promoters:", nps_counts['Promoter'] * 100, "%")
print("Passives:", nps_counts['Passive'] * 100, "%")
print("Detractors:", nps_counts['Detractor'] * 100, "%")
```

```python
# hyposis testing : 
# H1 customer attract to products that have 50% or high
# H0 customer not attract to products that have 50% or high
from scipy.stats import ttest_1samp
# Define the null hypothesis (H0) and alternative hypothesis (H1)
H1 = "Customers are attracted to products with a discounted percentage of 50% or higher"
H0 = "Customers are not attracted to products with a discounted percentage of 50% or higher"

# Set the significance level (alpha)
alpha = 0.05

# Perform one-sample t-test
t_statistic, p_value = ttest_1samp(df_new['discount_percentage'], 50)

# Print the results
print("Null Hypothesis (H0):", H0)
print("Alternative Hypothesis (H1):", H1)
print("Significance Level (alpha):", alpha)
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Determine whether to reject the null hypothesis
if p_value < alpha:
    print("Reject the null hypothesis. There is sufficient evidence to suggest that", H0)
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to reject", H1)
```
